# Train Test Split

```python
# Generate train/test split.
X_train, X_test, y_train, y_test = train_test_split(us_volumes_newEnvelopes.drop(columns = 'new_envelope_requests'),
                                                    us_volumes_newEnvelopes['new_envelope_requests'],
                                                    test_size = 0.2,
                                                    shuffle=False)
```

```python
# Check shape to confirm we did this properly.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Fit a linear model

```python
# Import statsmodels.
import statsmodels.api as sm
```

```python
# Add the intercept

X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)


# Confirm.
X_train.head()
```

```python
# statsmodels won't be able to handle missing values.
X_train.dropna(inplace=True)
y_train = y_train[X_train.index]

# This way we subset y_train to keep only indices from X_train.
```

```python
# Remember that, in statsmodels, we pass our data 
# in when we instantiate the model!

lm = sm.OLS(y_train, X_train)
```

```python
# Then we fit our model.
lm_results = lm.fit()
```

```python
# Display our summary!
print(lm_results.summary())
```

```python
# Generate predicted test values.
lm_results.predict(X_test)
```

```python
# Calculate R2 score.
r2_score(y_test, lm_results.predict(X_test))
```

```python
# Calculate RMSE.
mean_squared_error(y_test, lm_results.predict(X_test)) ** 0.5
```

```python
# Let's plot our predictions! 

# figure size
plt.figure(figsize=(20,10))

# Plot training data.
plt.plot(y_train.index, y_train.values, color = 'blue')

# Plot testing data.
plt.plot(y_test.index, y_test.values, color = 'orange')

# Plot predicted test values.
plt.plot(lm_results.predict(X_test), color = 'green')

# Set label.
plt.title(label = 'Forecasting Envelope Volumes', fontsize=24)

# Resize tick marks.
plt.xticks(fontsize=20)
plt.yticks(fontsize=20);
```

```python
us_volumes_newEnvelopes_predictions = pd.DataFrame({'new_envelope_requests': y_test.values,
                                                  'predicitons': lm_results.predict(X_test)})

us_volumes_newEnvelopes_predictions['difference'] = (us_volumes_newEnvelopes_predictions['predicitons'] - us_volumes_newEnvelopes_predictions['new_envelope_requests'])
us_volumes_newEnvelopes_predictions['%_difference'] = (us_volumes_newEnvelopes_predictions['new_envelope_requests'] - us_volumes_newEnvelopes_predictions['predicitons']) / us_volumes_newEnvelopes_predictions['new_envelope_requests']
us_volumes_newEnvelopes_predictions
```

```python
# figure size
plt.figure(figsize=(20,10))

# creat variables.
x_values = us_volumes_newEnvelopes_predictions[(us_volumes_newEnvelopes_predictions.index > '2023-07-01') & 
                                    (us_volumes_newEnvelopes_predictions.index < '2023-08-01')].index

new_envelop_values = us_volumes_newEnvelopes_predictions[(us_volumes_newEnvelopes_predictions.index > '2023-07-01') & 
                                    (us_volumes_newEnvelopes_predictions.index < '2023-08-01')]['new_envelope_requests'].values

prediction_values = us_volumes_newEnvelopes_predictions[(us_volumes_newEnvelopes_predictions.index > '2023-07-01') & 
                                    (us_volumes_newEnvelopes_predictions.index < '2023-08-01')]['predicitons'].values

difference_values = us_volumes_newEnvelopes_predictions[(us_volumes_newEnvelopes_predictions.index > '2023-07-01') & 
                                    (us_volumes_newEnvelopes_predictions.index < '2023-08-01')]['difference'].values


# Plot testing data.
plt.plot(x_values, new_envelop_values, color = 'orange', marker = 'o')
plt.plot(x_values, prediction_values, color = 'green', marker = 'o')

for i in np.arange(len(x_values) - 1):
    plt.annotate(f"{round(difference_values[i])}", xy =(x_values[i], prediction_values[i]), xytext=(5, -15), textcoords="offset pixels")
    

for i in x_values:
    plt.axvline(x = i, color = 'b', linestyle =  '--', linewidth = .5)

# Set label.
plt.title(label = 'Forecasting Envelope Volumes', fontsize=24)

# Resize tick marks.
plt.xticks(x_values, rotation = 90)
plt.yticks(fontsize=20);


```
